## Generate synthetic data

### Subtask:
Create synthetic data with noise based on a linear equation.


**Reasoning**:
Generate synthetic data with noise based on a linear equation as instructed.



```python
# 1. Define the true slope (m) and intercept (b)
m_true = 2
b_true = 5

# 2. Generate independent variable values (x)
np.random.seed(42) # for reproducibility
x = np.random.rand(100, 1) * 10 # Generate 100 random values between 0 and 10

# 3. Calculate the true dependent variable values (y_true)
y_true = m_true * x + b_true

# 4. Generate random noise
noise = np.random.randn(100, 1) * 2 # Add some noise

# 5. Create the observed dependent variable values (y)
y = y_true + noise

# Display the shapes of the generated data
print("Shape of x:", x.shape)
print("Shape of y_true:", y_true.shape)
print("Shape of noise:", noise.shape)
print("Shape of y:", y.shape)
```

## Implement gradient descent

### Subtask:
Define and implement the gradient descent algorithm to find the optimal slope (m) and intercept (b) for the linear regression model.


**Reasoning**:
Implement the hypothesis, cost function, and gradient descent algorithm as defined in the instructions.



```python
# 1. Define a function for the linear regression hypothesis
def hypothesis(x, m, b):
  """Calculates the predicted y values using the linear hypothesis."""
  return m * x + b

# 2. Define a function for the cost function (Mean Squared Error)
def cost_function(y_predicted, y_true):
  """Calculates the Mean Squared Error."""
  return np.mean((y_predicted - y_true)**2)

# 3. Define a function for the gradient descent algorithm
def gradient_descent(x, y, learning_rate, epochs):
  """Implements the gradient descent algorithm to find optimal m and b."""
  m = 0 # Initialize m
  b = 0 # Initialize b
  costs = [] # List to store costs at each epoch

  for epoch in range(epochs):
    y_predicted = hypothesis(x, m, b)
    cost = cost_function(y_predicted, y)
    costs.append(cost)

    # Calculate gradients
    m_gradient = -2 * np.mean(x * (y - y_predicted))
    b_gradient = -2 * np.mean(y - y_predicted)

    # Update m and b
    m = m - learning_rate * m_gradient
    b = b - learning_rate * b_gradient

  return m, b, costs

# 4. Call the gradient descent function
learning_rate = 0.01
epochs = 1000
m_learned, b_learned, costs = gradient_descent(x, y, learning_rate, epochs)

print("Learned m:", m_learned)
print("Learned b:", b_learned)
```

## Visualize results

### Subtask:
Plot the original data points, the fitted regression line, and the loss over epochs to show convergence.


**Reasoning**:
Create a scatter plot of the original data points and plot the fitted regression line. Add labels and a title to the plot.



```python
# 1. Create a scatter plot of the original data points
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Original Data', alpha=0.6)

# 2. Plot the fitted regression line
y_predicted = hypothesis(x, m_learned, b_learned)
plt.plot(x, y_predicted, color='red', label='Fitted Line')

# 3. Add labels and title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original Data and Fitted Linear Regression Line')
plt.legend()
plt.grid(True)
plt.show()
```

**Reasoning**:
Create a separate plot showing the cost over each epoch and add labels and a title to the plot. Then display the plot.



```python
# 4. Create a separate plot showing the cost over each epoch
plt.figure(figsize=(10, 6))
plt.plot(range(epochs), costs)

# 5. Add labels and title to the cost plot
plt.xlabel('Epoch')
plt.ylabel('Cost')
plt.title('Cost over Epochs')
plt.grid(True)

# 6. Display the plot
plt.show()
```
